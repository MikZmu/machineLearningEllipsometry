```python
import os

class locations():
    def __init__(self):
        """
        Initialize the locations class.
        """
        pass


    @staticmethod
    def get_notebook_parent_dir():
        """
        Get the path to the parent folder of the current Jupyter notebook.
        """
        notebook_path = os.getcwd()  # Current working directory of the notebook
        parent_dir = os.path.dirname(notebook_path)  # Parent directory
        return parent_dir

    @staticmethod
    def get_notebook_dir():
        """
        Get the path to the current Jupyter notebook.
        """
        notebook_path = os.getcwd()  # Current working directory of the notebook
        return notebook_path

    
    @staticmethod
    def get_models_dir():
        """
        Get the path to the models directory.
        """
        parent_dir = locations.get_notebook_dir()
        models_dir = os.path.join(parent_dir, 'models')
        return models_dir
    @staticmethod
    def get_data_dir():
        """
        Get the path to the data directory.
        """
        parent_dir = locations.get_notebook_dir()
        data_dir = os.path.join(parent_dir, 'datasets')
        return data_dir
    @staticmethod
    def get_scalers_dir():
        """
        Get the path to the scalers directory.
        """
        parent_dir = locations.get_notebook_dir()
        scalers_dir = os.path.join(parent_dir, 'scalers')
        return scalers_dir
    @staticmethod
    def list_models():
        """
        List all models in the models directory.
        """
        models_dir = locations.get_models_dir()
        return os.listdir(models_dir)
    @staticmethod
    def list_datasets():
        """
        List all datasets in the data directory.
        """
        data_dir = locations.get_data_dir()
        return os.listdir(data_dir)
    
    @staticmethod
    def list_scalers():
        """
        List all scalers in the scalers directory.
        """
        scalers_dir = locations.get_scalers_dir()
        return os.listdir(scalers_dir)
    
    
    @staticmethod
    def list_samples(dataset):
        data_dir = locations.get_data_dir()
        dataset = os.path.join(data_dir, dataset)
        return os.listdir(dataset)
    
    
    
    
    
    
    
    
```
